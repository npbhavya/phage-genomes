```python
#Import libraries
import matplotlib.pyplot as plt
import numpy as np
import csv
import re
import os
from os.path import exists
import seaborn as sns
import pandas as pd
```

```python
#########################################################
# Joining all samples assembly files to one dataframe
#########################################################
# plotting the coverage of the genomes compared the genome length
filepath="/home/nala0006/scratch/Bc-PhageSeq/github/crassphage/assembly"
assembler=["spades", "megahit", "flye", "unicycler"]
sample_names=["Bc01", "Bc02", "Bc03", "Bc04", "Bc05", "Bc06", "Bc07", "Bc08", "Bc09", "Bc10", "Bc11", "Bc12", "Bc13", "Bc14", "Bc15", "Bc16"]
result=pd.DataFrame()
for sample in sample_names:
    for f in assembler:
        filename=filepath+"/"+sample+"-assembly-stats_"+f+".tsv"
        print (filename)
        if (os.path.exists(filename)==True):
            assembly=pd.read_csv(filename, header=0)
            assembly.columns.values[2] = "Mean"
            assembly=assembly[assembly["Length_x"]>1000]
            assembly["SampleID"]=sample
            assembly["assemblers"]=f
            if (f=="spades" or f=="megahit"):
                assembly["sequencing"]="illumina"
            else:
                assembly["sequencing"]="nanopore"
            result = result.append(assembly)
        else:
            continue
display(result)
```

```python
#subsetting to visualize only nanopore assembly results 
nanopore = result['sequencing'].str.contains('nanopore', case=False, na=False)
result[nanopore]
```
